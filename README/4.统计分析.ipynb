```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("HTS DATA.csv")
df1.head()
```

```python
Y = df1.iloc[:,2]
fig = plt.figure(figsize=(5,3),dpi=100)
plt.hist(Y,bins=20)
ax = fig.add_subplot(111)
ax.set_xlabel(r'$bandgap$')
ax.set_ylabel(r'$frequency $')
plt.show()
```
