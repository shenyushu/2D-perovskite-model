```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.externals import joblib
from itertools import product 
```

```python
A1_des = pd.read_csv('A1_des.csv',index_col = 0)
A_des = pd.read_csv('A_des.csv',index_col = 0)  #将我自己的第一列设置为index
B_des = pd.read_csv('B_des.csv',index_col = 0)
X_des = pd.read_csv('X_des.csv',index_col = 0)
n_des = pd.read_csv('n_des.csv',index_col = 0)
ori_des = pd.read_csv('ori_des.csv',index_col = 0)
v = pd.read_csv('筛选出的变量名.csv')
v = np.array(v).tolist()
#v = v.values.tolist()
v = sum(v,[])  #略有一丝离谱啊，将表格里面的内容转为一维数组
v
```

```python
A1_sml = A1_des.index
A_sml = A_des.index
B_sml = B_des.index
X_sml = X_des.index
n_sml = n_des.index
ori_sml = ori_des.index
combination_sml = list(product(A1_sml,A_sml,B_sml,X_sml,n_sml,ori_sml))
combination_sml
```

```python
## 下面这个代码是嘉琪发给我的高通量计算的代码

#iloc与loc的差别在于，iloc是默认的顺序索引，而loc是实际名称的索引

model = joblib.load("Ridge.pkl")
results = []
#循环外的空集用来储存所有的数据
for a,b,c,d,e,f in combination_sml:
    result = []
    #循环内的空集用来储存每一次的结果，为什么要设置两个，是因为内部循环也需要把内容转换为值的列表进行储存
    series = n_des.loc[e,:].values.tolist() + ori_des.loc[f,:].values.tolist() + A1_des.loc[a,:].values.tolist() + A_des.loc[b,:].values.tolist() + B_des.loc[c,:].values.tolist()+ X_des.loc[d,:].values.tolist() 
    series = pd.DataFrame(series).T 
    #此时变成一个数组是为了添加行名列名
    series.columns =  n_des.columns.values.tolist() + ori_des.columns.values.tolist() + A1_des.columns.values.tolist() + A_des.columns.values.tolist() +B_des.columns.values.tolist() + X_des.columns.values.tolist() 
    #series.index = [a+b+c+d+e+f]
    series = series.loc[:,v]  
    #这一步是为了取出筛选之后的变量
    pred = model.predict(series.values)[0]
    result += [a,b,c,d,e,f,pred]
    result += series.values.tolist()[0]
    results.append(result)
results = pd.DataFrame(results)
results.columns = ["A1","A","B","X","n","ori","pred"]+v
#该处的v依旧是指筛选之后的变量名
results.iloc[:,6:] = results.iloc[:,6:].astype(float)
results.to_excel("results2.xlsx",index=None)
```
